```python
text='''2024 Enga landslide

Article
Talk
Read
Edit
View history

Tools
Coordinates: 5°22′26″S 143°23′19″E
From Wikipedia, the free encyclopedia
2024 Enga landslide
2024 Enga landslide is located in Papua New Guinea2024 Enga landslide
Date	24 May 2024
Time	c. 03:00 PGT (UTC+10:00)
Location	Maip Muritaka Rural LLG, Enga Province, Papua New Guinea
Coordinates	5°22′26″S 143°23′19″E[citation needed]
Cause	Gold mining and/or heavy rains (suspected)[1][2]
Deaths	11 (confirmed)
160–2,000 (estimated)
Non-fatal injuries	17
Missing	2,500+[3][4]
On 24 May 2024, a landslide occurred in Mulitaka, Papua New Guinea. Estimates of the number of deaths have ranged from 160 to 2,000.[5] A week later, 11 bodies had been recovered,[4][6] but many more people were missing, including about 2,500 in the villages of Kaokalam and Tulipana alone.

Background
Papua New Guinea has regularly experienced fatal landslides resulting from its mountainous terrain, weather, climate, poverty, poor land use practices and government mismanagement. In 2024, the country saw intense rainfall and flooding,[2] and in April, 14 people died in a landslide, while 21 died in a landslide one month earlier.[7]

Cause
On 18 May, a Mw 4.5 earthquake occurred 105 km (65 mi) west of where the landslide happened. It struck 126.2 km (78.4 mi) below the surface[8] and has been suggested as a possible cause of the landslide.[1] However, the Red Cross has stated that there was no indication that the earthquake caused the landslide, instead attributing it to either gold mining or heavy rain.[2]

Papua New Guinea's prime minister James Marape has blamed climate change for the disaster.[9]

Impact
The landslide occurred at approximately 03:00 PGT on 24 May 2024 (17:00 UTC on 23 May),[10] after a large amount of debris was dislodged from the limestone slopes of Mount Mungalo.[11][2] Some survivors said that they heard two loud cracks around one to two hours before the disaster, alerting them to possible danger and enabling their escape.[12] The landslide destroyed six villages in Maip Muritaka Rural LLG.[2] In Kaokalam village alone, dozens of houses were destroyed and an estimated 500 people were buried.[4] It blocked a highway near the Porgera Gold Mine, and destroyed 150 metres (490 ft) of the main highway leading into Kaokalam, causing concerns over the supply of fuel and goods.[13][1][14] The mine's management said that its regular operations were unaffected and that fuel at the site was enough to sustain the facility for 40 days.[15] A road construction crew working on the highway was also buried in the landslide.[16]

Approximately 2,000 people were buried in Tulipana village.[3] The Highlands Highway connecting the area to Port Moresby, the capital, was also blocked.[17] Gardens which provided food for the village and its three water streams were buried and destroyed.[18] An official of the International Organization for Migration estimated that the area covered by the landslide was the equivalent of between "three and four football fields".[19] Later estimates put the area covered by the landslide at 600 metres (2,000 ft) long and 90,000 square metres (970,000 sq ft).[20] The debris was estimated to be 6 metres (20 ft) to 8 metres (26 ft) deep.[21]

Conflicting estimates of casualties emerged following the disaster, which was attributed to the lack of updated population statistics since the last national census in 2000[22] and the fact that the affected area also hosted pubs catering to miners from other villages[23] as well as people displaced by intertribal fighting.[24] On 27 May, it was estimated that 2,000 people may have died by the landslide in Yambali village alone.[25] These numbers were not confirmed by government officials.[19] On 29 May, councillor and chair of the Mulitaka Disaster Committee Jaman Yandam said that more than 160 people had died, citing village leaders who had conducted a head count.[23] Over 4,000 people were directly affected.[26] Five bodies and a leg of a sixth victim were recovered.[11] At least 17 people were injured[27] and four others were rescued, while 1,182 homes were reported to have been destroyed or buried.[28][29] More than 5,000 pigs, 100 stores, and five vehicles were also believed to have been buried,[30] as well as two health centres.[31] An additional 250 homes were condemned due to unstable ground conditions, displacing about 1,250 people.[11]

Aftermath
My wife and I were deeply shocked and saddened to learn of the devastating landslide in Enga, and the tragic loss of so many lives, homes and food gardens. [...] I have faith that your communities will come together to support the survivors and the recovery in these heartbreaking circumstances.[32]

Charles III, King of Papua New Guinea, 2024

Prime Minister James Marape announced that the Papua New Guinea Defence Force was sent to the scene to conduct relief work, recover bodies, and rebuild destroyed infrastructure. Police, medics, engineers, and United Nations personnel were also dispatched, while some locals acted as first responders.[33] By dusk, rescuers at the scene utilized a mechanical digger and tools in an attempt to locate survivors.[34] The Papua New Guinea government formally requested aid from the United Nations and other countries on 26 May.[35] A state of emergency was declared in the affected region,[36] with authorities saying that nearly 7,900 people living near the disaster site were at risk from further landslides.[37] Condolences were sent by Governor-General Bob Dadae and King Charles III.[38][39]

International humanitarian agency CARE and the Papua New Guinea Red Cross Society said that they were assessing the situation.[40][41][42] CARE later said around 4,000 people were in need of humanitarian aid following the disaster, including individuals displaced during intertribal clashes in February 2024.[26] Australia, France, India and the United States said they were prepared to assist in humanitarian efforts,[19][43][44][45] with Australia subsequently providing AU$2.5 million ($1.7 million) and a disaster response team.[35] The Australian Defence Force and the Queensland Fire and Emergency Services were also deployed to assist in relief efforts.[46][47] The Indian government announced relief assistance of US$1 million on 28 May, reaffirming the two countries' ties under the Forum for India-Pacific Islands Cooperation (FIPIC).[48][49] The European Union released 150,000 euros in aid,[50] while USAID provided.[51] On 28 May, the New Zealand Government pledged NZ$1.5 million in relief support[52] and announced plans to deploy a New Zealand Defence Force C-130 aircraft carrying relief supplies and technical expertise.[53] The provincial governments of Hela and Southern Highlands pledged K1 million in aid each, while ICTSI South Pacific and Newmont also pledged K1 million respectively.[54][55]

Rescue work was hampered by large rocks and fallen trees,[56] with the Red Cross estimating that it would take up to two days for humanitarian assistance to arrive.[2] Police Commander Martin Kelei reported that recovery efforts further were complicated by the potential danger of additional landslides caused by removing debris, while Coordinator for CARE Australia Justine McMahon stated that instability and shifting of the disaster site and forecasts of future rain could lead to further landslides threatening rescuers and aid staff.[57] UN migration agency of Papua New Guinea Serhan Aktoprak reported that the landslide engulfed and blocked three streams located near the villages affected, soaking the ground even more and increasing the risk of future landslides in the recovery area.[58] According to ABC News, only helicopters could access Kaokalam.[59] Clearing of the blocked highway was also delayed due to opposition from some of the relatives of the victims, who called for the site to be preserved in situ as a memorial.[60]

Humanitarian convoys were deployed by road from the provincial capital Wabag toward the disaster zone beginning on 25 May and were escorted by the Papua New Guinea Defence Force to protect the convoys as they passed through Tambitanis, where intertribal clashes resulted in 8 deaths and the burning of 35 homes and businesses. Defense Minister Billy Joseph and PNG National Disaster Centre director Laso Mana visited Wabag on 26 May to assess recovery efforts.[11] The Porgera mine also pledged to provide mechanical diggers to help in rescue and road-clearing operations.[61] The National Disaster Centre released K500,000 in aid to the Enga provincial government.[62] Heavy equipment arrived at the site on 26 May,[63] with more expected to be delivered by the Papua New Guinea Defence Force from Lae as early as 28 May.[64] Aid groups expressed concerns over the landslide's effects on youth, who make up 40% of the affected population. At least nine children were orphaned by the landslide.[65] Authorities also announced plans to quarantine the disaster zone to prevent the spread of disease.[66]

A minute of silence was held at the National Parliament of Papua New Guinea to commemorate the victims.[23] During a parliamentary session on 30 May 2024, Speaker Job Pomat reprimanded MPs for a muted response to the disaster, which occurred amid plans to file a no-confidence motion against Marape.[67] On 31 May, Marape visited the site of the landslide[68] and pledged K20 million towards reconstruction efforts.[69]'''
```

```python
# Run on raw text

inputs = tokenizer(text, return_tensors="pt", padding="longest", truncation=True).to("cuda")
outputs = model(**inputs)
logits = outputs.logits.squeeze(-1).float().detach().cpu().numpy()
score = logits.item()
result = {
    "text": text,
    "score": score,
    "int_score": int(round(max(0, min(score, 5))))
}

print(result["score"])
```
